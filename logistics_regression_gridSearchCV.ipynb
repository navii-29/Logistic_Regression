```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 1000,n_features = 10 , n_classes = 2)
```

```python
# no need for scaling already scaled data
```

```python
data = pd.DataFrame(X)
```

```python
display(data,type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression() # c is 1/lambda of ridge lasso parameter 
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# prediciton probabiliy 
model.predict_proba(X_test)[0]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test , y_pred)
print("accuracy score is", score)
```

```python
cm = confusion_matrix(y_test , y_pred)
print("confusion matrix"'\n', cm )
```

```python
report = classification_report(y_test,y_pred)
print(report)
```

# Hyper parameter tuning using Grid Search CV

```python
tuned_model = LogisticRegression()
```

```python
# adding all possible values of parameter in the grid for tuned_model
```

```python
# Fix 1: Increase max_iter to stop ConvergenceWarnings
tuned_model = LogisticRegression(max_iter=10000, random_state=42)

# Fix 2: Use a list of dictionaries to map valid penalties to valid solvers
params = [
    # Config 1: l1 penalty (supported by liblinear and saga)
    {
        'penalty': ['l1'],
        'solver': ['liblinear', 'saga'],
        'C': [0.1, 1, 10, 100, 0.01]
    },
    
    # Config 2: l2 penalty (supported by all solvers)
    {
        'penalty': ['l2'],
        'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
        'C': [0.1, 1, 10, 100, 0.01]
    },
    
    # Config 3: elasticnet (only supported by saga, and requires l1_ratio)
    {
        'penalty': ['elasticnet'],
        'solver': ['saga'],
        'C': [0.1, 1, 10, 100, 0.01],
        'l1_ratio': [0.5] # elasticnet needs this parameter (0.5 = 50% l1, 50% l2)
    }
]

# Now run GridSearchCV with these corrected params
from sklearn.model_selection import StratifiedKFold, GridSearchCV
cv = StratifiedKFold(n_splits=5)

grid = GridSearchCV(tuned_model, param_grid=params, n_jobs=-1, scoring='accuracy', cv=cv)
grid.fit(X_train, y_train)
```
